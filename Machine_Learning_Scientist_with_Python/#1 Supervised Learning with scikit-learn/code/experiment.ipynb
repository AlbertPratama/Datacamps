## R^2 0r R-Squared



#### `R2 =  [ (nΣxy – (Σx)(Σy)) / (√nΣx2-(Σx)2 * √nΣy2-(Σy)2) ]2`

### calculate each metric that we need to use in the R2


### Calculate R-Squared

#### R2 =  [ (nΣxy – (Σx)(Σy)) / (√nΣx2-(Σx)2 * √nΣy2-(Σy)2) ]2
#### R2 =  [ (8*(2169) – (72)(223)) / (√8*(818)-(72)2 * √8*(6447)-(223)2) ]2
#### R2 =  0.6686

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data X dan Y
# Mengubah bentuk menjadi kolom
x = np.array([3, 5, 5, 7, 9, 12, 14, 17]).reshape(-1, 1)
y = np.array([22, 24, 28, 20, 28, 31, 37, 33])

# Membuat model regresi linear
model = LinearRegression()
model.fit(x, y)

# Memprediksi nilai Y
y_pred = model.predict(x)

# Menghitung R²
r_squared = r2_score(y, y_pred)

print(f'Nilai R²: {r_squared}')
```
