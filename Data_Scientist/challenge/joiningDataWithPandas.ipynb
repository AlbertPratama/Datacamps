```python
import pandas as pd

# Create the 'books' DataFrame
books = pd.DataFrame({
    'book_id': [142000671, 261103288, 451529308],
    'title': ['Of Mice and Men', 'The Hobbit', 'Little Women'],
    'num_pages': [352, 277, 449],
    'author': ['John Steinbeck', 'J.R.R. Tolkien', 'Louisa May Alcott']
})

# Create the 'reviews' DataFrame
reviews = pd.DataFrame({
    'rating': [4.42, 4.49, 4.24],
    'book_id': [142000671, 261103288, 140283331],
    'title': ['Of Mice and Men', 'The Hobbit', 'Lord of the Flies'],
    'reviewer_id': [3, 1, 2]
})
books
```

```python
reviews
```

```python
# Merge the 'books' and 'reviews' DataFrames on the 'book_id' column
books_reviews = books.merge(reviews, on='book_id')

# Filter the 'books' DataFrame to include only the books that have reviews
top_books = books[books['book_id'].isin(books_reviews['book_id'])]

# Display the 'top_books' DataFrame
print(top_books)
```
