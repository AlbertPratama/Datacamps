```python
import sqlite3
import pandas as pd

# Membaca data dari CSV ke DataFrame
review = pd.read_csv('../datasets/reviews.csv')

# Nama file database SQLite
db_file = 'review.db'

# Membuat koneksi ke database SQLite
conn = sqlite3.connect(db_file)

# Menulis DataFrame ke tabel dalam database SQLite
review.to_sql('reviews', conn, if_exists='replace', index=False)

# Menutup koneksi
conn.close()

print("DataFrame telah berhasil diubah menjadi database SQLite.")
```

```python
import sqlite3
import pandas as pd

# Buat koneksi ke database
conn = sqlite3.connect('review.db')

# Eksekusi query dan simpan hasil ke DataFrame
df = pd.read_sql_query("SELECT * FROM reviews", conn)

# Tampilkan DataFrame
print(df)

# Tutup koneksi
conn.close()
```
